##Extracting Numbers from string:


```python
Data_set['Unique ID'] = Data_set['Unique ID'].str.extract('(\d+)')
Data_set
```

##Indexing:


```python
Data_set.set_index('Unique ID' ,inplace= True)
Data_set
```

##Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
lebelencoder = LabelEncoder()
Data_set['SellerG'] = lebelencoder.fit_transform(Data_set['SellerG'])
Data_set['Suburb'] = lebelencoder.fit_transform(Data_set['Suburb'])
Data_set['Address'] = lebelencoder.fit_transform(Data_set['Address'])
Data_set['Type'] = lebelencoder.fit_transform(Data_set['Type'])
Data_set['Method'] = lebelencoder.fit_transform(Data_set['Method'])
Data_set['Date'] = lebelencoder.fit_transform(Data_set['Date'])
Data_set['CouncilArea'] = lebelencoder.fit_transform(Data_set['CouncilArea'])
Data_set['Regionname'] = lebelencoder.fit_transform(Data_set['Regionname'])
Data_set



```

##Dealing with NaN Values:


```python
# Data_set.isna().sum()
Data_set = Data_set.interpolate()
Data_set
```

```python
Data_set = Data_set.dropna()
Data_set
```
