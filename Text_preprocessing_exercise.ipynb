```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import re
```

##Load Dataset Using pandas

```python
Data_set = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tripadvisor_hotel_reviews.csv')
# Data_set.size
```

##Rating Distribution

```python
def temp(rating):
    if(rating <=2 ):
        rating = 0
    elif(rating== 3):
        rating = 1
    else:
        rating = 2
    return rating
            

Data_set.Rating = Data_set.Rating.apply(temp )
Data_set     
```

##Dealing with missing values

```python
Data_set.isna().sum()    # Here we can clearly see that there is no NAN values in our dataset. So no need to Deal with it 
```

##Lower All the Text

```python
def lower_text(temp):
    temp.lower()
    return temp

Data_set.Review = Data_set.Review.apply( lower_text)
Data_set           
```

##Remove Hyperlinks

```python
def remove_hyperlinks(temp):
    temp  = re.sub(r'^https?:\/\/.*[\r\n]*', '', temp)
    return temp
Data_set.Review = Data_set.Review.apply( remove_hyperlinks)
Data_set    
```

##Substitute all the non alphabetical characters

```python
def remove_nonalphabetical(temp):
    temp  = re.sub("[^a-zA-Z]+", ' ', temp)
    return temp
Data_set.Review = Data_set.Review.apply( remove_nonalphabetical )
Data_set  
```

##Remove Spaces

```python
def remove_spaces(temp):
    temp  = re.sub(' +',' ', temp)
    return temp
Data_set.Review = Data_set.Review.apply( remove_spaces)
Data_set
```

```python
sw_nltk = stopwords.words('english')
# print(sw_nltk)
```

```python
def Remove_Stopwords(x):
  x = [word for word in x if word.lower() not in sw_nltk]
  return x
Data_set.Review = Data_set.Review.apply(Remove_Stopwords)
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
```

```python
def Stemming(x):
  return [porter.stem(y) for y in x]
 

Data_set.Review = Data_set.Review.apply(Stemming)
```
