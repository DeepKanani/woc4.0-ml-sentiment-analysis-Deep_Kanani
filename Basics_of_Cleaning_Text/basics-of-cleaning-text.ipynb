# **Load Dataset Using pandas**

```python
Data_set = pd.read_csv('/kaggle/input/trip-advisor-hotel-reviews/tripadvisor_hotel_reviews.csv')
# Data_set.size
```

# **Rating Distribution**

```python
def temp(rating):
    if(rating <=2 ):
        rating = 0
    elif(rating== 3):
        rating = 1
    else:
        rating = 2
    return rating
            

Data_set.Rating = Data_set.Rating.apply(temp )
Data_set        
```

**Dealing with missing values**

```python
Data_set.isna().sum()    # Here we can clearly see that there is no NAN values in our dataset. So no need to Deal with it 
```

# **Lower All the Text**

```python
def lower_text(temp):
    temp.lower()
    return temp

Data_set.Review = Data_set.Review.apply( lower_text)
Data_set           
```

# **Remove Hyperlinks**

```python
def remove_hyperlinks(temp):
    temp  = re.sub(r'^https?:\/\/.*[\r\n]*', '', temp)
    return temp
Data_set.Review = Data_set.Review.apply( remove_hyperlinks)
Data_set    
```

**Substitute all the non alphabetical characters**

```python
def remove_nonalphabetical(temp):
    temp  = re.sub("[^a-zA-Z]+", ' ', temp)
    return temp
Data_set.Review = Data_set.Review.apply( remove_nonalphabetical )
Data_set  
```

# **Remove Spaces**

```python
def remove_spaces(temp):
    temp  = re.sub(' +',' ', temp)
    return temp
Data_set.Review = Data_set.Review.apply( remove_spaces)
Data_set   
```
